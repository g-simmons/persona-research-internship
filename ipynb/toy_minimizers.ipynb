```python
def himmelblau(x, y): 
   return (((x**2) + y - 11)**2) + ((x + (y**2) - 7)**2)

himmelblau(3, 2)
```

```python
import numpy as np
grid = np.linspace(-6, 6, 100)
```

```python
points = np.meshgrid(grid, grid)
values = himmelblau(points[0], points[1])
```

```python
# minimizers = np.array(list(zip(grid,grid)))[values < 50]
# print(len(minimizers))

minimizers = np.array(list(zip(points[0].flatten(), points[1].flatten())))[values.flatten() < 10]
print(len(minimizers))
```

```python
minimizers
```

```python
minimizers_sample_idx=np.random.choice(np.arange(len(minimizers)), 30)
```

```python
minimizers_sample = minimizers[minimizers_sample_idx]
```

```python
import matplotlib.pyplot as plt

X, Y = np.meshgrid(grid, grid)
Z = himmelblau(X, Y)

plt.contour(X, Y, Z, levels=40)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Himmelblau Function Contour Plot')

plt.scatter(minimizers_sample[:,0], minimizers_sample[:,1], color='red', marker='*')

plt.show()


```

```python
minimizers_sample_function_values = himmelblau(minimizers_sample[:,0], minimizers_sample[:,1])
```

```python
#save the points and function values to a file
```

```python
import pandas as pd

minimizers_sample_df = pd.DataFrame(minimizers_sample, columns=['x', 'y'])
minimizers_sample_df['function_value'] = minimizers_sample_function_values

minimizers_sample_df.to_csv('../DualAlpha/data/minimizers_sample.csv', index=False)
```
